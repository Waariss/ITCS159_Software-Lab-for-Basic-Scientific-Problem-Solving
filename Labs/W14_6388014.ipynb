```python
#Ex11
#mask = Circle area
import numpy as np
test_image = img.copy()
lx , ly , lz = test_image.shape
X, Y = np.ogrid [0:lx , 0: ly]
mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 4;
test_image [mask]=0;
io.imshow(test_image)
io.show()
```

What is the result of blurring image? Can you blur more?


```python
from skimage import filters
blurred_img = filters.gaussian(img , sigma= 50 , multichannel = True)
io.imshow(blurred_img)
io.show()
```

```python
#Ex17

alpha = 0.15
shapened_img = img + alpha * (img-blurred_img)
io.imshow(shapened_img.astype(int))
io.show()
io.imshow(img)
io.show()
```

What is the result of sharpening image? What does the equation means?

Equation make the image more sharpness and erase the blur.
alpha = Sharpness
(img-blurred_img) = Cut the blur area out
alpha * (img-blurred_img) = Imge have more sharpness
